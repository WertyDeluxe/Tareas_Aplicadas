1. $x^5-6x^3+2x=2$

*imagen* comprobatoria de la funcion en Geogebra"image.png

```python
# Función f(x) y su derivada
f = lambda x: x**5 - 6*x**3 + 2*x - 2
fprime = lambda x: 5*x**4 - 18*x**2 + 2

# Graficar la función
x_vals = np.linspace(-3, 3, 400)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title("Gráfica de f(x) = x^5 - 6x^3 + 2x - 2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()

# Método de Newton-Raphson
def newton_raphson(f, fprime, x0, tol, maxiter):
    x = x0
    for i in range(maxiter):
        fx = f(x)
        if abs(fx) < tol:  # Si la función en x es suficientemente pequeña, terminamos
            return x
        fp = fprime(x)
        if abs(fp) < 1e-12:  # Evitar dividir por cero si la derivada es muy pequeña
            print('Error: la derivada es demasiado pequeña')
            return None
        x = x - fx / fp
    return x

# Parámetros
x0 = -3  # Aproximación inicial ajustada visualmente desde la gráfica
tol = 1e-4
maxiter = 10000

# Resultado
root = newton_raphson(f, fprime, x0, tol, maxiter)
if root is not None:
    print(f"La raíz de la ecuación es: {root}")
else:
    print("No se encontró una raíz válida.")
```

2. $senx=x^2$

image.png

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función f(x) = sin(x) - x^2 y su derivada
def f1(x):
    return np.sin(x) - x**2

def df1(x):
    return np.cos(x) - 2*x

# Método de Newton
def metodo_newton(f, df, x0, tol=1e-4, max_iter=100):
    for _ in range(max_iter):
        x1 = x0 - f(x0)/df(x0)
        if abs(x1 - x0) < tol:
            return x1
        x0 = x1
    return x0

# Graficamos para estimar una solución inicial
x = np.linspace(-1, 1, 400)
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, x**2, label='x^2')
plt.title('sin(x) = x^2')
plt.grid(True)
plt.legend()
plt.show()

# Usamos como valor inicial: x0 ≈ 0
x0 = 0.0
solucion = metodo_newton(f1, df1, x0)
print(f"Solución para sin(x) = x^2: x ≈ {solucion}")
```

```python
# Definimos la función f(t) = cos(t) - t^2 y su derivada
def f2(t):
    return np.cos(t) - t**2

def df2(t):
    return -np.sin(t) - 2*t

# Método de Newton
def metodo_newton(f, df, t0, tol=1e-4, max_iter=100):
    for _ in range(max_iter):
        t1 = t0 - f(t0)/df(t0)
        if abs(t1 - t0) < tol:
            return t1
        t0 = t1
    return t0

# Graficamos para estimar una solución inicial
t = np.linspace(-1, 2, 400)
plt.plot(t, np.cos(t), label='cos(t)')
plt.plot(t, t**2, label='t^2')
plt.title('cos(t) = t^2')
plt.grid(True)
plt.legend()
plt.show()

# Usamos como valor inicial: t0 ≈ 0.7
t0 = 0.7
solucion = metodo_newton(f2, df2, t0)
print(f"Solución para cos(t) = t^2: t ≈ {solucion}")
```

```python
# Definimos la función f(x) = 2sin(x) - sqrt(x+1) y su derivada
def f3(x):
    return 2*np.sin(x) - np.sqrt(x + 1)

def df3(x):
    return 2*np.cos(x) - 0.5 / np.sqrt(x + 1)

# Método de Newton
def metodo_newton(f, df, x0, tol=1e-4, max_iter=100):
    for _ in range(max_iter):
        x1 = x0 - f(x0)/df(x0)
        if abs(x1 - x0) < tol:
            return x1
        x0 = x1
    return x0

# Graficamos para estimar una solución inicial
x = np.linspace(-0.9, 2, 400)
plt.plot(x, 2*np.sin(x), label='2sin(x)')
plt.plot(x, np.sqrt(x + 1), label='sqrt(x + 1)')
plt.title('2sin(x) = sqrt(x+1)')
plt.grid(True)
plt.legend()
plt.show()

# Usamos como valor inicial: x0 ≈ 1.0
x0 = 1.0
solucion = metodo_newton(f3, df3, x0)
print(f"Solución para 2sin(x) = sqrt(x+1): x ≈ {solucion}")
```

II. Use Newton’s method to approximate $\sqrt[3]{2}$ with an error less than 0.00005.  Use $x^n-k=0$


```python
def f(x):
    return x**3 - 2

def df(x):
    return 3 * x**2

def metodo_newton(f, df, x0, tol=5e-5, max_iter=100):
    for i in range(max_iter):
        x1 = x0 - f(x0) / df(x0)
        print(f"Iteración {i+1}: x ≈ {x1}")
        if abs(x1 - x0) < tol:
            return x1
        x0 = x1
    return x0

# Usamos como valor inicial una suposición cercana: x0 = 1.5
x0 = 1.5
solucion = metodo_newton(f, df, x0)

print(f"\nAproximación de la raíz cúbica de 2: {solucion}")
```

##Solucion I
## **1. $f(x) = \frac{x}{x+1}$**

###  Paso 1: Análisis previo (sin cálculo)

* **Dominio**: $x \ne -1$ (no se puede dividir entre 0).
* **Rango**: Todos los reales excepto $y = 1$ (asíntota horizontal).
* **Simetría**:
  $f(-x) = \frac{-x}{-x + 1} \ne f(x)$ → no es par.
  $f(-x) = \frac{-x}{1 - x} \ne -f(x)$ → no es impar.
* **Intersección con el eje x**: $f(x) = 0 \Rightarrow x = 0$ → punto (0, 0)
* **Intersección con el eje y**: Igual que arriba, (0, 0)

###  Paso 2: Análisis con cálculo

* **Primera derivada**:

  $$
  f(x) = \frac{x}{x+1} \Rightarrow f'(x) = \frac{1(x+1) - x(1)}{(x+1)^2} = \frac{1}{(x+1)^2} > 0
  $$

  ⇒ La función siempre es creciente (excepto en $x = -1$, donde no está definida).

* **Segunda derivada**:

  $$
  f''(x) = \frac{-2}{(x+1)^3} \Rightarrow
  \begin{cases}
  < 0 & \text{si } x > -1 \\
  > 0 & \text{si } x < -1
  \end{cases}
  $$

  ⇒ Concavidad hacia abajo a la derecha de $x = -1$, hacia arriba a la izquierda.

* **Asíntotas**:

  * Vertical: $x = -1$
  * Horizontal: $y = 1$ (cuando $x \to \infty$, $f(x) \to 1$)

###  Paso 3 y 4: Bosquejo

* Curva creciente.
* Asíntota vertical en $x = -1$
* Asíntota horizontal en $y = 1$
* Pasa por (0, 0)

---

## **2. $g(x) = \frac{|x| + x}{2}(3x + 2)$**

###  Paso 1: Análisis previo

* **Dominio**: Todos los reales.

* **Simetría**:

  * Para $x \geq 0$: $|x| + x = 2x$
  * Para $x < 0$: $|x| + x = 0$
    ⇒ Entonces:

  $$
  g(x) =
  \begin{cases}
  x(3x + 2) = 3x^2 + 2x & x \geq 0 \\
  0 & x < 0
  \end{cases}
  $$

  ⇒ No es par ni impar.

* **Intersecciones**:

  * Eje y: $g(0) = 0$
  * Eje x: Solución de $x(3x + 2) = 0$ → $x = 0$ y $x = -\frac{2}{3}$ (pero esta última solo válida si $x \geq 0$)

###  Paso 2: Análisis con cálculo

* **Función por tramos**:

  * Para $x < 0$: constante 0.
  * Para $x \geq 0$: derivamos $3x^2 + 2x$:

    $$
    g'(x) = 6x + 2
    $$

    ⇒ Crece siempre para $x > 0$

* **Concavidad**:

  $$
  g''(x) = 6 > 0 \Rightarrow \text{Cóncava hacia arriba}
  $$

* **Punto de no derivabilidad**: en $x = 0$, hay un quiebre (punto angular).

###  Paso 3 y 4: Bosquejo

* Constante 0 para $x < 0$.
* Parabólica creciente desde $x = 0$ hacia la derecha.
* Punto (0, 0) como unión entre ambas ramas.

---

## **3. $f(x) = \frac{x^2}{x^2 - 1}$**

###  Paso 1: Análisis previo

* **Dominio**: $x \ne \pm 1$
* **Simetría**: $f(-x) = f(x)$ → función **par**.
* **Intersección con eje y**: $x = 0 \Rightarrow f(0) = 0$

###  Paso 2: Cálculo

* **Primera derivada**:

  $$
  f'(x) = \frac{(2x)(x^2 - 1) - x^2(2x)}{(x^2 - 1)^2} = \frac{2x(x^2 - 1 - x^2)}{(x^2 - 1)^2} = \frac{-2x}{(x^2 - 1)^2}
  $$

  ⇒ $f'(x) < 0$ si $x > 0$, $f'(x) > 0$ si $x < 0$ → máximo local en $x = 0$


* **Asíntotas**:

  * Verticales: $x = \pm 1$
  * Horizontal: $\lim_{x \to \infty} \frac{x^2}{x^2 - 1} = 1$

###  Paso 3 y 4: Bosquejo

* Par, simétrica respecto al eje y.
* Asíntotas en $x = \pm 1$
* Tiende a 1 cuando $x \to \pm \infty$
* Tiene un máximo en (0, 0)

## **4. $h(x) = \frac{5.235x^3 - 1.245}{7.126x - 3.141}$**

###  Paso 1: Análisis previo

* **Dominio**: $x \ne \frac{3.141}{7.126} \approx 0.441$
* **Simetría**: No es par ni impar (coeficientes arbitrarios).
* **Intersección eje y**: $ x = 0 \Rightarrow h(0) = \frac{-1.245}{-3.141} \approx 0.396$

###  Paso 2: Cálculo

* **Primera derivada** (regla del cociente):

Sea:

$$
u = 5.235x^3 - 1.245,\quad v = 7.126x - 3.141
$$

$$
h'(x) = \frac{u'v - uv'}{v^2}
= \frac{(15.705x^2)(v) - (u)(7.126)}{(v)^2}
$$


* **Asíntota vertical**: $x = 0.441$
* **Asíntota oblicua**: Como el grado del numerador es mayor que el del denominador, hay asíntota oblicua. Usa división larga para encontrarla.

###  Paso 3 y 4: Bosquejo

* Curva racional con una asíntota vertical en $x \approx 0.441$
* Se comporta como una parábola hacia los extremos (porque el numerador es cúbico y el denominador lineal)
* Cruza eje y cerca de 0.396

---

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000)
x_safe1 = x[x != -1]
x_safe3 = x[(x != -1) & (x != 1)]
x_safe4 = x[x != 3.141/7.126]

# f1
f1 = x_safe1 / (x_safe1 + 1)

# g(x)
g = np.where(x >= 0, x * (3*x + 2), 0)

# f3
f3 = x_safe3**2 / (x_safe3**2 - 1)

# h(x)
h = (5.235 * x_safe4**3 - 1.245) / (7.126 * x_safe4 - 3.141)

# Plot
plt.figure(figsize=(12, 10))

# f1
plt.subplot(2, 2, 1)
plt.plot(x_safe1, f1, label='f(x) = x / (x + 1)')
plt.axvline(-1, color='red', linestyle='--')
plt.axhline(1, color='gray', linestyle=':')
plt.title('f(x) = x / (x + 1)')
plt.grid()
plt.legend()

# g
plt.subplot(2, 2, 2)
plt.plot(x, g, label='g(x)', color='green')
plt.title('g(x) = [(|x| + x)/2] * (3x + 2)')
plt.grid()
plt.legend()

# f3
plt.subplot(2, 2, 3)
plt.plot(x_safe3, f3, label='f(x) = x^2 / (x^2 - 1)', color='purple')
plt.axvline(-1, color='red', linestyle='--')
plt.axvline(1, color='red', linestyle='--')
plt.axhline(1, color='gray', linestyle=':')
plt.title('f(x) = x^2 / (x^2 - 1)')
plt.grid()
plt.legend()

# h
plt.subplot(2, 2, 4)
plt.plot(x_safe4, h, label='h(x)', color='orange')
plt.axvline(3.141/7.126, color='red', linestyle='--')
plt.title('h(x) = (5.235x^3 - 1.245) / (7.126x - 3.141)')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```


##Solucion III
### Paso 1: Análisis antes del cálculo

**(a) Dominio y rango**  
- $f$ es continua en todas partes $\implies$ dominio: $\mathbb{R}$ (todos los reales).  
- No hay información explícita sobre el rango, pero se infiere que incluye al menos $[-3, \infty)$ por $f(2)=-3$ y $f(6)=1$ con $f$ creciente para $x > 2$.

**(b) Simetría**  
- No hay información sobre simetría (no es par ni impar, ya que $f(2)=-3 \neq f(-2)$ y $f(2) \neq -f(-2)$).  

**(c) Intersecciones**  
- Intersección en $y$: No dada.  
- Intersección en $x$: No dada (pero podría existir en $x < 2$ o $x > 6$).  

---

### Paso 2: Análisis con cálculo

**(a) Primera derivada ($f'$)**  
- $f'(2) = 0$ (punto crítico en $x=2$).  
- $f'(x) \neq 0$ para $x \neq 2$ $\implies$ no hay otros puntos críticos.  
- $f'(6) = 3 > 0$ $\implies$ $f$ es creciente en $x=6$.  
- Como $f'(x)$ no cambia de signo excepto en $x=2$, y $f'(6)>0$, inferimos que $f'(x) > 0$ para $x > 2$.  

**(b) Puntos críticos**  
- $x=2$ es un **mínimo local** porque:  
  - $f''(x) > 0$ en $(2, 6)$ $\implies$ cóncava hacia arriba.  
  - $f'(2) = 0$ y $f$ es creciente después ($f'(x) > 0$).  

**(c) Segunda derivada ($f''$)**  
- $f''(x) > 0$ en $(2, 6)$ $\implies$ cóncava hacia arriba en este intervalo.  
- $f''(x) < 0$ para $x > 6$ $\implies$ cóncava hacia abajo.  
- $f''(6) = 0$ $\implies$ posible punto de inflexión en $x=6$.  

**(d) Asíntotas**  
- No hay asíntotas verticales (continua en $\mathbb{R}$).  
- No hay información sobre asíntotas horizontales u oblicuas.  

---

### Paso 3: Puntos clave
- $(2, -3)$: Mínimo local.  
- $(6, 1)$: Punto de inflexión (cambia de cóncava a convexa).  

---

### Paso 4: Bosquejo de la gráfica

1. **Para $x < 2$**:  
   - $f'(x) \neq 0$ y no hay máximos/mínimos.  
   - Comportamiento no especificado, pero asumimos que $f$ decrece hacia $x=2$ (para que $x=2$ sea mínimo).  

2. **En $x=2$**:  
   - Mínimo local en $(2, -3)$.  

3. **Para $2 < x < 6$**:  
   - $f$ crece ($f'(x) > 0$) y es cóncava hacia arriba ($f''(x) > 0$).  

4. **En $x=6$**:  
   - Punto de inflexión en $(6, 1)$.  
   - $f'(6) = 3$ (pendiente positiva).  

5. **Para $x > 6$**:  
   - $f$ sigue creciendo ($f'(x) > 0$) pero es cóncava hacia abajo ($f''(x) < 0$).  

---

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.where(x < 2, (x - 2)**2 - 3, np.where(x <= 6, 0.25*(x - 2)**2 - 3, 3*(x - 6) + 1))

x = np.linspace(-2, 10, 400)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$f(x)$', color='blue')
plt.scatter([2, 6], [-3, 1], color='red', zorder=5)
plt.annotate('Mínimo en (2, -3)', (2, -3), xytext=(0, -15), textcoords='offset points')
plt.annotate('Punto de inflexion en (6, 1)', (6, 1), xytext=(0, 10), textcoords='offset points')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.title("Bosquejo de $f(x)$")
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
plt.show()
```
