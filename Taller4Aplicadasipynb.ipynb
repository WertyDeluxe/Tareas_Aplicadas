# 1) Serie Fibonacci:
Escribe un programa en Python que genere los primeros n números
de la serie de Fibonacci y visualízalos en un gráfico de líneas utilizando Matplotlib. El
usuario debe poder elegir el número n

```python
import matplotlib.pyplot as plt

def fibonacci(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]

    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[i-1] + fib[i-2])
    return fib

n = int(input("Ingrese el valor de n: "))
fib_series = fibonacci(n)

print(fib_series)

plt.plot(fib_series, marker='o', linestyle='-', color='b', label="Serie Fibonacci")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.title("Gráfico de la Serie de Fibonacci")
plt.legend()
plt.grid()
plt.show()
```

# 4) Gráfica de funciones trigonométricas:
Escribe un programa que grafique las
funciones seno, coseno y tangente en el mismo plano para valores entre 0 y 2π.

```python
# Generar valores de x entre 0 y 2π
x = np.linspace(0, 2 * np.pi, 1000)

# Calcular funciones trigonometricas
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

# Configurar la grafica
plt.figure(figsize=(10, 5))
plt.plot(x, y_sin, label='Seno', color='blue')
plt.plot(x, y_cos, label='Coseno', color='red')
plt.plot(x, y_tan, label='Tangente', color='green')

# Evitar valores extremos en la tangente
plt.ylim(-2, 2)

# Añadir leyenda y etiquetas
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.title("Funciones Trigonometricas")
plt.xlabel("x (radianes)")
plt.ylabel("Valor de la función")
plt.grid()
plt.show()
```

# 5) Raíces cuadradas:
Grafique la función 𝑦 = √𝑥 junto con la línea 𝑦 = 𝑥 en el mismo
gráfico para valores de x entre 0 y 10

```python
# Generar valores de x entre 0 y 10
x = np.linspace(0, 10, 100)

# Calcular las funciones
y_sqrt = np.sqrt(x)
y_line = x

# Configurar la grafica
plt.figure(figsize=(8, 5))
plt.plot(x, y_sqrt, label=r'$y = \sqrt{x}$', color='blue')
plt.plot(x, y_line, label=r'$y = x$', color='red', linestyle='dashed')

# Añadir leyenda y etiquetas
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.title("Grafica de Raices Cuadradas y la funcion y = x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```
