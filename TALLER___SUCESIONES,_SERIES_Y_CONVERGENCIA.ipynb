Diviendo los terminos por n $(para _n\gt 0)$, obtenemos: $$-\frac{1}{n}\le \frac{sin(n)}{n}\le \frac{1}{n}$$ lo que justifica la desigualdad dada.

Aplicamos el teorema del estrangulamiento, observamos que los extremos de la desigualdad, $-\frac{1}{n}$, y $\frac{1}{n}$, tienden a 0 cuando n ---> ‚àû
$$\lim_{n \to \infty}- \frac{1}{n}=0 ,  \lim_{n \to \infty} \frac{1}{n}=0$$

El primer t√©rmino es $a=7.5m$

La raz√≥n es $r=0.75$

$$S_n\approx {\text{ln}}(n)+\gamma$$

Donde:

$n$=n√∫mero de t√©rminos

$\gamma \approx 0.577$ es la **constante de Euler-Mascheroni**

Podemos despejar $n$ de esta expresi√≥n cuando $S_n>5$:

$${\text{ln}}(n)+0.577>5$$

Restamos 0.577 en ambos lados:

$${\text{ln}}(n)>4.423$$

Aplicamos la funci√≥n exponencial:
$$n>e^{4.423}\approx 83.1$$

Conclusi√≥n: Se necesitan aproximadamente 84 t√©rminos para que la suma sea mayor que 5.

Ahora repetimos el proceso con $S_n>10$:

$${\text{ln}}(n)+0.577>10$$

Restamos 0.577:

$${\text{ln}}(n)>9.423$$

Aplicamos la funci√≥n exponencial:

$$n>e^{9.423}\approx 12300$$

Conclusi√≥n: Se necesitan aproximadamente 12,300 t√©rminos para que la suma sea mayor que 10.

>Para que la suma parcial sea mayor que 5, se necesitan alrededor de 84 t√©rminos.

>Para que la suma parcial sea mayor que 10, se necesitan alrededor de 12,300 t√©rminos.
